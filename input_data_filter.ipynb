```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Analyzing content for one file
# Reading the source file
df = pd.read_csv('/Users/biswanathpatro/Study/machine learning/Proj Ramin/Facial Expression Analysis/data/b_emphasis_datapoints.txt',sep=' ')
# Reading the target file
df_t = pd.read_csv('/Users/biswanathpatro/Study/machine learning/Proj Ramin/Facial Expression Analysis/data/a_affirmative_targets.txt',names = ['Target'])
df = pd.concat([df,df_t],axis=1)
df_t = df[df['Target'] == 1]
```

```python
def medianTreatment():
    myseries1 = df_t.iloc[:,3::3].quantile(0.25) == 0
    list1 = myseries1[myseries1].index.tolist() 

    myseries2 = df_t.iloc[:,3::3].quantile(0.50) == 0
    list2 = myseries2[myseries2].index.tolist()
    
    for i in range(0,len(list1)):
        str1 = list1[i]
        df_t[str1] = df_t[str1].fillna(0.0).astype(int)
        df_t.loc[df_t[str1]==0, str1]=(df_t[str1].median())
        
    for i in range(0,len(list2)):
        str2 = list2[i]
        print(str2)
        df_t.loc[df_t[str2]==0, str2]=(df_t[str2].mean()*2)        
        
medianTreatment()
```
